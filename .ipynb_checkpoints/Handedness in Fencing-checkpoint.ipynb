# In-Depth Statistical Analysis

So it certainly looks as though there are more left-handers than there should be in the top 100, and they tend to be higher ranked within the top 100 than their right-handed peers. Let's do a statistical analysis to determine whether this is purely down to chance.

This section is the most technical, and assumes a reasonable grasp of bayesian inference. Skip ahead to the <strong>Conclusions</strong> section if you'd rather avoid the details!

### Initial Analysis - Binomial Model

The top 100 can be considered as a random sample from the general population. The sample is without replacement, but because 100 is so small compared to the world population, we may treat it as a sample with replacement. For each sample datum, we are interested in their handedness, which can take one of only two values. Sampling with replacement when there are two possible outcomes suggests a binomial distribution.

So we let $X$ denote the random variable being the number of left-handers in a sample of 100 top fencers, and suppose $X$ is binomially distributed with $n=100$ samples and some probability $q$ of a given fencer being left-handed. If handedness has no effect on fencing, then $q$ will be the proportion of the total population which is left-handed, namely 10.6%; if left-handers do have an advantage in fencing, then $q$ will be higher.

First, we compute the likelihood function of $q$ given the data that $X=36$.

```python
# likelihood of q given L=36 is probability that L=36 given various different values of q

def likelihood(q):
    return stats.binom.pmf(36, n=100, p=q)
```

```python
# plot the likelihood function

# sample 1000 different possible values of q
qvec = np.linspace(0, 1, 1000)

# compute the likelihood of each of these values of q
lvec = likelihood(qvec)

# plot the likelihood for each of those values of q
plt.plot(qvec, lvec)
plt.title("Likelihood function of q given that 36 of the top 100 fencers are left-handed")
plt.xlabel("q")
plt.ylabel("L(q|X=36)")
```

We see that the likelihood is very close to 0 except between about 0.2 and about 0.5; the posterior distribution of $q$ is found by multiplying this likelihood function by the prior distribution for $q$ (and normalising), so we see that unless our prior distribution has very susbtantial probability mass outside the interval from 0.2 to 0.5, the posterior distribution will have most of its probability mass in this interval.

In other words, unless we have a *very* strong prior belief about the value of $q$, the evidence will lead us to conclude that $q$ is very likely to be between 0.2 and 0.5. If handedness had no effect on fencing ability, then $q$ would be around 10.6%, or 0.106, as this is the proportion of left-handers in the general population. So the evidence initially seems conclusive that being left-handed *does* confer a significant advantage in fencing, unless we adopt a very restrictive prior to the contrary.

Let's now consider some different prior distributions for $q$ and see what we conclude from the evidence about the value of $q$ under these priors. We will use beta-distributions for the priors, as is standard when modelling a probability. So we assume $q$ follows a beta distribution, which has two parameters, $\alpha$ and $\beta$.

The mode of a beta distribution is given by
$$\frac{\alpha-1}{\alpha + \beta -2},$$
at least so long as $\alpha$ and $\beta$ are both greater than 1. It is reasonable to make the mode of our prior distribution equal to the population parameter of 0.106. This gives
$$\begin{align*} \frac{\alpha-1}{\alpha + \beta -2}&=0.106\\ \alpha-1 &= 0.106(\alpha + \beta -2)\\ 0.894\alpha - 1 &= 0.106\beta - 0.212\\ \beta &= \frac{0.894\alpha - 0.788}{0.106}.\end{align*}$$
So we can pick a value for $\alpha$, and then take $\beta$ according to this formula to obtain a beta distribution with mode 0.106. Larger values of $\alpha$ will give a more peaked distribution, while smaller values will allow more spread.

```python
# define a function to compute beta from alpha, such that the mode will be 0.0106

def beta(alpha):
    return (0.894 * alpha - 0.788) / 0.106
```

```python
fig, ax = plt.subplots(1,1)
for alpha in [2, 4, 6]:
    ax.plot(qvec, stats.beta.pdf(qvec, a=alpha, b=beta(alpha)))
plt.xlabel("q")
plt.ylabel("P(q|alpha)")
plt.title("A selection of beta prior distributions for $q$, all with mode 0.106, and different peakedness")
plt.legend(["alpha = 2", "alpha = 4", "alpha = 6"])
plt.show()
```

These seem like not unreasonable priors to take; the alpha=6 (green) curve, for instance, models a prior belief that left-handers should be expected to make up no more than about 20% of top fencers; given that 20% is already twice the population average, this allows for a fairly substantial bias, while assuming that no bias is the most likely case.

```python
# define a function to compute the posterior distributions for each of the above priors,
# using the likelihood function from before

# first compute an unnormalised version, as we will need to integrate that to normalise it
def unnormalised_posterior(alpha, q):
    return likelihood(q) * stats.beta.pdf(q, a=alpha, b=beta(alpha))

# now normalise
def posterior(alpha, q):
    return unnormalised_posterior(alpha, q) / integrate(lambda x: unnormalised_posterior(alpha, x), 0, 1)[0]
```

```python
# plot the posteriors for each of the above priors

fig, ax = plt.subplots(1,1)
for alpha in [2, 4, 6]:
    ax.plot(qvec, posterior(alpha, qvec))
plt.xlabel("q")
plt.ylabel("P(q | alpha, X=36)")
plt.title("Posterior distributions for q, for a range of priors")
plt.legend(["alpha = 2", "alpha = 4", "alpha = 6"])
plt.show()
```

So for various prior beliefs about how likely an advantage was for left-handers, these graphs show the updated beliefs in light of the evidence. Note that even with the most restrictive prior, when alpha=6, the posterior has virtually all of its probability mass between 0.2 and 0.4, whereas the prior had almost all of its probability mass below 0.2.

So even if we were initially fairly convinced that any advantage left-handers may have in fencing is small, in light of the evidence of 36 left-handers out of the top 100, we should be confident that in fact there is a significant advantage, such that left-handers should be expected to take up at least 20% of top fencers.

How significant is this result? Let's compute some likelihood ratios, comparing these prior hypotheses to the null hypothesis that no bias exists (so $q=0.106$).

We can find the likelihood of the null hypothesis by simply evaluating the likelihood of $q$ function at $q=1$. For the other priors, we will need to integrate over the prior:
$$L(\alpha\mid X=36) = \int_0^1 L(q\mid X=36)P(q\mid \alpha)\;\mathrm{d}q,$$
where $L(q\mid X=36)$ is the likelihood-of-$q$ function, and $P(q\mid \alpha)$ is the prior distribution of $q$ for the given value of $\alpha$.

```python
# compute the likelihood of the null hypothesis given the data

null_likelihood = likelihood(0.106)
```

```python
# define a function to compute likelihoods of beta prior hypotheses given various values of alpha

def beta_likelihood(alpha):
    def integrand(q):
        return likelihood(q) * stats.beta.pdf(q, a=alpha, b=beta(alpha))
    return integrate(integrand, 0, 1)[0]
```

```python
# compute likelihood ratios to compare each of the beta priors with the null hypothesis

for alpha in [2, 4, 6]:
    print("The likelihood ratio of alpha being %d against the null is %.2f\n"
          % (alpha, beta_likelihood(alpha) / null_likelihood))
```

A likelihood ratio of 1 would mean the evidence was indecisive either way; enormous likelihood ratios like these mean that we can be virtually certain the null hypothesis is false.

This does not mean any of the beta priors is correct; merely that they are *far* more likely than the null. Each of the beta priors is shifted by the evidence to give a posterior substantially to the right of the prior, so in light of the evidence our beliefs about $q$ should be broadly in line with the posterior distributions obtained, and we certainly should not believe $q$ is anywhere close to 0.106.

So we can confidently conclude that the large proportion of left-handers in the top 100 fencers is not down to chance alone. There is a genuine and substantial bias in favour of left-handers.

### Further Analysis - Wallenius Model

So far we've looked at the fact that 36 out of the top 100 are left-handed. Comparing this to the general population convinced us that elite fencers are significantly more likely to be left-handed than we would expect by chance, so there does seem to be a substantial advantage for left-handers. However, it is possible that left-handers are more common among fencers *of all abilities* than in the general population, and so this could explain the observed number of left-handers, without left-handedness actually being an advantage.

Anecdotally, I can assure you that far fewer than 36% of all fencers are left-handed, but I do not have actual data on the prevalence of left-handedness among the fencing population at large. As such, our analysis so far does not conclusively show that left-handedness is an advantage in fencing, though it certainly suggests that.

For further analysis then, we can consider how the left-handers are distributed among the top 100. We have already observed from the boxplot that there seem to be more left-handers with particularly high numbers of ranking points, so let's explore this further. We will take it as given that 36 out of the top 100 fencers are left-handed, and examine how likely it is to have them arranged within the top 100 as they are, given that proportion.

To perform this analysis, imagine having the top 100 in a room and calling 50 of them (say) out, one at a time. We know that there are 36 left-handers and 64 right-handers in the room initially, so we can model how many left-handers we are likely to get in our random 50. If left-handedness has no effect on fencing performance, then calling them out in order of their rankings should have no effect on this experiment: if their handedness does not affect their performance, then ranking order is as good as random order as far as handedness goes. On the other hand, if left-handers are advantaged at fencing, then calling out the top 50 in ranking order should give us more left-handers than we would get calling out 50 at random.

This sort of situation is modelled by the pretentiously named <a href="https://en.wikipedia.org/wiki/Wallenius%27_noncentral_hypergeometric_distribution" target="_blank"> Wallenius' noncentral hypergeometric distribution</a>. We have $N$ items (100 fencers) in a room, and draw $n$ of them (50, say) out, one at a time and without replacement. We know that $m$ out of the $N$ are type-1, and the rest are type-2 (36 out of the 100 are left-handed, the rest are right-handed); if our selection method is biased in favour of type-1 items by a ratio $w$, then Wallenius' distribution tells us, for each $k$, the probability of having $k$ type-1 items in our sample of $n$.

So the null hypothesis (that handedness has no effect on fencing performance) corresponds to a Wallenius distribution with weight ratio $w=1$, whereas a bias in favour of left-handers will correspond to $w>1$. The value of $w$ can be thought of as how much stronger left-handers are than right-handers.

Unfortunately, the SciPy stats package does not currently have Wallenius' distribution implemented (though hopefully it will soon - I'm discussing this with the devs!). So we'll have to hard code the (rather unpleasant) probability mass function:
$$P(k\mid N,n,m,w) = \binom{m}{k}\binom{N-m}{n-k}\int_0^1 \left(1-t^{w/D}\right)^k \left(1-t^{1/D}\right)^{n-k}\;\mathrm{d}t,$$
where
$$D=w(m-k) + ((N-m) - (n-k)).$$

```python
# define the Wallenius pmf:

from scipy.special import binom as choose # binomial coefficient function

def wallenius_pmf(k, N, m, n, w):
    """m is the number of type-1 objects, N the total number (so N-m is the number of type-2);
    w is the weight ratio; it represents how much more likely a type-1 object is to be drawn than a type-2.
    n is the number of items drawn (one at a time, without replacement)
    The support is from max(0, n-m_2) to min(n, m_1) inclusive.
    For given N, m, n, w, the function will say for each k how probable it is to have k-many type-1's in the draw of n"""
    D = w * (m - k) + ((N - m) - (n - k))
    def integrand(t):
        return ((1 - t ** (w / D)) ** k) * ((1 - t ** (1 / D)) ** (n - k))
    return choose(m, k) * choose(N - m, n - k) * integrate(integrand, 0, 1)[0]
```

```python
# for us, N=100 (we're looking at the top 100 fencers), and m=36 (we know there are 36 left-handers in the top 100)

N = 100
m = 36
```

We now need to choose a value of $n$ (how many to sample from the top 100), and look at the data to find the value of $k$ we get (how many left-handers are in the sample). We will try a few different values of $n$, to see how it affects the results; small values of $n$ will be more susceptible to random variation, while larger values will be less able to detect to genuine bias in favour of left-handers. Increasing $n$ reduces the sensitivity while increasing the specificity.

For each value of $n$ we use, we will find the value of $k$ from the dataframe, then compute the likelihood function for $w$ given that data. Since we will do this for several values of $n$, we might as well write our likelihood function to take $n$ as an argument.

```python
# define a function of n and w which for fixed n gives the likelihood function of w given the evidence from the top n fencers

def wallenius_likelihood(n, w):
    # look up the k-value for the given n
    k = (df.iloc[:n]['hand'] == 'L').sum()

    return wallenius_pmf(k, N, m, n, w)
```

To plot the likelihood function, we should technically plot over all positive real values of $w$. However, since $w$ is the bias in favour of left-handers, it is unlikely that $w$ is extremely large - after all, if the bias in favour of left-handers were *really* extreme, then we'd expect to see even more than 36 left-handers in the top 100. So we'll plot the likelihood function for $w<10$, say, and then see if we need to adjust the range of $w$ values.

```python
# plot the likelihood of w for several different values of n

# take 1000 points for w between 0 and 10 (might need adjusting )
wvec = np.linspace(0, 10, 1000)

# the y-axis scale will be different for each value of n, so best to plot on separate axes
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(10,10), sharex=True)
subplot_list = [ax1, ax2, ax3, ax4, ax5, ax6]
for subplot, n in enumerate([10, 20, 30, 40, 50, 60]):
    lvec = np.zeros(len(wvec))
    # wallenius_likelihood takes scalar arguments, but wvec is an array, so we need a loop to evaluate the likelihood
    for index, w in enumerate(wvec):
        lvec[index] = wallenius_likelihood(n, w)
    subplot_list[subplot].plot(wvec, lvec)
    subplot_list[subplot].set_title("n = %d" %n)
fig.suptitle("Likelihood of w, for a range of different values of n")
plt.show()
```

There is a general trend as $n$ increases of the likelihood function becoming more peaked. This is to be expected, because larger samples are more likely to approach the mean. For each of the different values of $n$ considered, the maximum of the likelihood function occurs around $w=2$, so under a broad range of reasonable priors, we would expect our posterior to suggest $w$ is around 2.

However, note that in each case, the likelihood is already starting to pick up around $w=1$. So actually it's not clear that this gives conclusive evidence that $w>1$. To explore this more, let's consider some reasonable priors and compute likelihood ratios. Henceforth we will use $n=30$, as this seems fairly representative of the likelihood after the broad spread coming from low $n$ values has settled down.

What sort of prior distribution should we use for $w$? A reasonable candidate is a <a href="https://en.wikipedia.org/wiki/Beta_prime_distribution" target="_blank">beta-prime distribution</a>; like the beta distribution we used as a prior for $q$ in the binomial analysis, the beta prime distribution has shape parameters we can use to fine-tune it. Unlike the beta distribution, however, the beta prime distribution is supported on the positive reals, making it a suitable distribution for $w$, which could in principle be any positive real.

A beta-prime distribution has parameters $\alpha$ and $\beta$; its mode is
$$\frac{\alpha -1}{\beta + 1},$$
so to have a mode of 1 (corresponding to the prior belief that handedness has no effect on fencing ability) we want
$$\begin{align*} \frac{\alpha -1}{\beta + 1} &=1\\ \alpha -1 &= \beta + 1\\ \beta &= \alpha - 2 \end{align*}.$$
So we are free to choose a value of $\alpha$, and then use $\beta=\alpha-2$.

```python
# given alpha, find beta to make the mode of the beta-prime distribution 1

def beta(alpha):
    return alpha - 2
```

```python
# plot a selection of beta-prime prior distributions for w

fig, ax = plt.subplots(1,1)
for alpha in [3, 6, 12]:
    ax.plot(wvec, stats.betaprime.pdf(wvec, a=alpha, b=beta(alpha)))
plt.xlabel("w")
plt.ylabel("P(w|alpha)")
plt.title("A selection of beta-prime prior distributions for $w$, all with mode 1, and different peakedness")
plt.legend(["alpha = 3", "alpha = 6", "alpha = 12"])
plt.show()
```

We see that larger values of $\alpha$ give a more peaked distribution, corresponding to a firmer belief that any bias due to handedness is small.

Another possible prior distribution, the "null hypothesis," is that $w=1$ exactly, with no other value possible. We shall also consider this hypothesis.

So we shall compute the likelihood of the hypotheses $w=1$, $\alpha=3$, $\alpha=6$, and $\alpha=12$ given the evidence from the top 100 fencers when $n=30$.

For the null hypothesis, this is straightforward: we just need to evaluate the wallenius_likelihood function at $n=30$ and $w=1$. For the other hypotheses, with beta-prime distributions, we must integrate over all values of $w$:
$$L(\alpha \mid n) = \int_0^\infty L(w \mid n)P(w)\;\mathrm{d}w,$$
where $L(w\mid n)$ is the wallenius_likelihood function we defined above, and $P(w)$ is the beta-prime prior distribution of $w$.

```python
# define a function to compute the likelihood of a beta-prime hypothesis given the data

def beta_prime_likelihood(n, alpha):
    def integrand(w):
        return wallenius_likelihood(n, w) * stats.betaprime.pdf(w, a=alpha, b=beta(alpha))
    return integrate(integrand, 0, np.inf)[0]
```

```python
# compute the likelihood of the null hypothesis, with n=30, w=1

n = 30
null_likelihood = wallenius_likelihood(n, 1)
```

```python
# compute likelihood ratios to compare each of the beta-prime priors with the null hypothesis

for alpha in [3, 6, 12]:
    print("The likelihood ratio of alpha being %d against the null is %.2f\n"
          % (alpha, beta_prime_likelihood(n, alpha) / null_likelihood))
```

So the evidence provides less than 2-to-1 evidence in favour of the beta-prime prior against the null hypothesis.

So looking purely at how the left-handers are distributed within the top 100, we have only weak evidence in favour of a left-handed bias - despite the fact that the top 3 are all left-handed! This is probably partly because 100 is not a very large number; we might be able to get stronger evidence by combining data from other age, sex, and weapon categories - though it is also possible that the strength of any bias towards left-handers varies by category, so combining categories could confuse the issue.

## Conclusions

The proportion of left-handed fencers in the world top 100 is far larger than can reasonably be explained by chance, so we conclude that being left-handed does confer an advantage in competitive fencing. Without data on the frequency of left-handedness within the fencing population at large (as opposed to the general population), it is difficult to say quite how much of an advantage there is, but we can reasonably expect a group of elite fencers to contain upwards of 20% left-handers, and likely close to 30%, compared with only 10.6% in the general population.

Not only are there more left-handers in the top 100 than one would expect if there were no advantage, but also within the top 100 the left-handers are more likely to be highly ranked than their right-handed peers. However, this effect, and the sample size, are too small to say with much confidence that left-handers do markedly better within the top 100.

So it seems that being left-handed is almost certainly an advantage when it comes to getting into the top 100, and may be a slight advantage when it comes to rising up higher within the top 100.
